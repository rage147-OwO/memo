# 승하차 상위 5개 정류장

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

bus_data=pd.read_excel(r"C:\Users\Hoseo\Documents\GitHub\2023DataOnAir\day5\(첨부2)내포순환버스 승하차이용현황_230225~230604_호서대.xlsx")
gettingInData=bus_data[["시간대","승차정류장명"]]
#dropna
gettingInData.dropna(inplace=True)
s=gettingInData.groupby(["승차정류장명"]).승차정류장명.count()
s.sort_values(ascending=False,inplace=True)
slicedData=s[0:3].to_dict()

wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",background_color="white",width=1000,height=1000,max_words=100,max_font_size=300).generate_from_frequencies(slicedData)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

bus_data=pd.read_excel(r"C:\Users\Hoseo\Documents\GitHub\2023DataOnAir\day5\(첨부2)내포순환버스 승하차이용현황_230225~230604_호서대.xlsx")
gettingInData=bus_data[["시간","승차정류장명"]]
gettingInData.dropna(inplace=True)

#출근시간대 상위 정류장
morningData=gettingInData[(gettingInData["시간"]>=8) & (gettingInData["시간"]<=10)]
morningData=morningData.groupby(["승차정류장명"]).승차정류장명.count()



#퇴근시간대 상위 정류장
eveningData=gettingInData[(gettingInData["시간"]>=17) & (gettingInData["시간"]<=19)]
eveningData=eveningData.groupby(["승차정류장명"]).승차정류장명.count()

concatedData=morningData+eveningData
concatedData.sort_values(ascending=False,inplace=True)
slicedData=concatedData[0:3]

#visualization
wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf",background_color="white",width=1000,height=1000,max_words=100,max_font_size=300).generate_from_frequencies(slicedData)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()










```
